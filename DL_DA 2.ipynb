```python
#importing essential libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Sigmoid function
```

```python
def sigmoid(x):
 return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
#tanh
```

```python
def tanh(x):
 return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
#ReLU
```

```python
def RELU(x):
 x1=[]
 for i in x:
    if i<0:
      x1.append(0)
    else:
       x1.append(i)
 return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```

```python
#Leaky ReLU
```

```python
from matplotlib import pyplot
pyplot.style.use('ggplot')
pyplot.figure(figsize=(10,5))
```

```python
def Leaky_ReLU(x):
 if x>0:
    return x
 else: 
    return 0.01*x
```

```python
# define a series of inputs
input_series = [x for x in range(-19, 19)]
# calculate outputs for our inputs
output_series = [Leaky_ReLU(x) for x in input_series]
# line plot of raw inputs to rectified outputs
pyplot.plot(input_series, output_series)
plt.title('Activation Function :Leaky RELU')
pyplot.show()
```

```python
#Exponential ReLU
```

```python
#exponential ReLU
def elu(arr,alpha):
 a=[]
 for x in arr:
   if x >= 0:
      a.append(x)
   else:
      a.append(alpha * (np.exp(x)-1))
 return a
```

```python
z=elu(x,1.0)
plt.plot(x,z)
plt.title('Activation function: Exponential ReLU')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Conv2D
import skimage.measure
```

```python
from PIL import Image
image=Image.open('/content/drive/MyDrive/pqr.png')
plt.imshow(image)
```

```python
rsize = image.resize((np.array((15,15))).astype(int))
image = np.asarray(rsize)
image.shape
```

```python
x=image.reshape(1,15,15,3)
```

```python
def Convolution(x,kernel_size,strides=1, padding='valid'):
  model=Sequential()
  model.add(Conv2D(1, kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(15,15,3)))
  yhat = model.predict(x)
  y=yhat.reshape(yhat.shape[1],yhat.shape[2])
  filtered_image = Image.fromarray(y)
  return filtered_image
```

```python
#3x3 filter
image_1=Convolution(x,(3,3)) #filter size 3*3
plt.imshow(image_1)
entropy = skimage.measure.shannon_entropy(image_1)
print("Entropy:",entropy)
```

```python
#5x5 filter
image_2=Convolution(x,(5,5))
plt.imshow(image_2)
entropy = skimage.measure.shannon_entropy(image_2)
print("Entropy:",entropy)
```

```python
#3x3 stride=1
image_3=Convolution(x,(3,3),1,'same')
plt.imshow(image_3)
entropy = skimage.measure.shannon_entropy(image_3)
print("Entropy:",entropy)
```

```python
#3x3 stride=2
image_4=Convolution(x,(3,3),2) 
plt.imshow(image_4)
entropy = skimage.measure.shannon_entropy(image_4)
print("Entropy:",entropy)
```

```python
#5x5 stride=1
image_5=Convolution(x,(5,5),1,'same')
plt.imshow(image_5)
entropy = skimage.measure.shannon_entropy(image_5)
print("Entropy:",entropy)
```

```python
#5x5 stride= 2
image_6=Convolution(x,(5,5),2) 
plt.imshow(image_6)
entropy = skimage.measure.shannon_entropy(image_6)
print("Entropy:",entropy)
```
